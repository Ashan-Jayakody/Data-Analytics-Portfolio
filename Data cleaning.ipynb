```python
def calculateSalaryRanges(salary):
    if salary <= 100000:
        return '100K and Below'
    if salary > 100000 and salary <= 200000:
        return '100K-200K'
    if salary > 200000 and salary <= 300000:
        return '200K-300K'
    if salary > 300000 and salary <= 400000:
        return '300K-400K'
    if salary > 400000 and salary <= 500000:
        return '400K-500K'
    if salary > 500000 and salary <= 600000:
        return '500K-600K'
    if salary > 600000 and salary <= 700000:
        return '600K-700K'
    if salary > 700000 and salary <= 800000:
        return '700K-800K'
    return'other'
```

```python
df['salary_ranges'] = df['salary_in_usd'].apply(calculateSalaryRanges)
```
